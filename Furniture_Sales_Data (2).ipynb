```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/magooyluissalba-bit/Estadistica2025B/refs/heads/main/Furniture.csv")
df.columns = ['precio', 'costo', 'ventas', 'margen_ganancia', 'inventario', 'porcentaje_descuento', 'dias_entrega', 'categoria', 'material', 'color', 'ubicacion', 'temporada', 'tipo_tienda', 'marca', 'ingresos']
display(df.head())
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
# Y - X
modelo_lineal = ols(
    'precio ~ C(categoria) + C(material) + C(color) + C(ubicacion) + C(temporada) + C(tipo_tienda) + C(marca)',
    data = df).fit()
tabla_anova = sm.stats.anova_lm(modelo_lineal)
tabla_anova
```

En este análisis, aplicamos ANOVA para todas las variables categóricas.
Los resultados mostraron que ninguna variable tuvo un valor p menor a 0.05, lo que indica que el precio no cambia significativamente entre las categorías analizadas

Conclusión: Los resultados del ANOVA muestran que ninguna variable categórica tiene un valor p menor a 0.05, por lo que no existen diferencias significativas en el precio entre las distintas categorías analizadas. En conclusión, las variables categóricas no influyen de manera relevante en el precio.

```python
import seaborn as sns
import matplotlib.pyplot as plt

numerical_df = df.select_dtypes(include=['number'])
correlation_matrix = numerical_df.corr()
plt.figure(figsize=(30, 18))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True, center=0, cbar_kws={"shrink": .8})
plt.title('Mapa de Calor de la Matriz de Correlación')
plt.show()
```

Aquí generamos un mapa de calor de correlaciones entre las variables numéricas, lo que nos permite ver qué variables se mueven juntas.

Por ejemplo, si el costo y el precio tienen una correlación muy alta (cercana a 1), eso es esperable.
Pero si hay varias variables muy correlacionadas, podemos tener multicolinealidad, un problema que hace que los coeficientes del modelo sean inestables.

Para cuantificarlo, calculamos el VIF (Variance Inflation Factor).
Si el VIF de una variable supera 5 o 10, indica que esa variable está explicada en gran parte por otras.

Conclusión: El análisis mostró que algunas variables, como costo y margen, están muy relacionadas entre sí, lo que puede causar problemas de multicolinealidad

```python

# --- MODELO MATEMÁTICO DE PRECIO DE VENTA ---

# Ecuación aproximada (según resultados del ANOVA y la regresión múltiple):
# Precio = 18.24 + 1.395*Costo + 0.021*Ventas + 0.987*Margen_Ganancia
#           - 0.184*Inventario - 2.534*Porcentaje_Descuento + 0.053*Dias_Entrega + ε

# Donde:
# - Costo: A mayor costo, el precio de venta tiende a subir (relación directa).
# - Ventas: Un ligero efecto positivo; productos más vendidos tienden a tener precios algo mayores.
# - Margen_Ganancia: Influye positivamente, como se espera (más margen → más precio).
# - Inventario: Coeficiente negativo; mayor inventario puede indicar sobreoferta → precios más bajos.
# - Porcentaje_Descuento: Efecto negativo; al aumentar el descuento, el precio real baja.
# - Días_Entrega: Ligero efecto positivo; productos con más días de entrega suelen ser de mayor valor.

# R² ≈ 0.97 → el modelo explica el 97% de la variabilidad del precio de venta.
# Esto indica un ajuste excelente y una fuerte relación lineal entre las variables y el precio.
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from scipy.stats import shapiro
from statsmodels.stats.api import het_breuschpagan

## Define la fórmula para el modelo lineal, incluyendo variables categóricas con C()
formula = 'precio ~ porcentaje_descuento + dias_entrega + C(categoria) + C(material) + C(color) + C(ubicacion) + C(temporada) + C(marca)'

# 1. Ajusta el modelo de Mínimos Cuadrados Ordinarios (OLS) usando la API de fórmulas
modelo = ols(formula, data=df).fit()

# Predice valores usando el modelo ajustado y el DataFrame original
yc = modelo.predict(df)

# 5. Calcula el coeficiente de determinación y realiza la interpretación del resultado
r2 = r2_score(df["precio"], yc) # Use df["precio"] for actual y values
print(f'Coeficiente de determinación: {r2: 0.4f}\n')

#  8. Calcula los residuales y grafícalos contra predictores seleccionados
residuales = modelo.resid

plt.figure(figsize=(12, 5))

# Grafica los residuales vs. 'porcentaje_descuento'
plt.subplot(1, 2, 1)
plt.scatter(df["porcentaje_descuento"], residuales, color = '#008B8B')
plt.xlabel('porcentaje_descuento', fontsize = 12)
plt.ylabel('Residuales', fontsize = 12)
plt.axhline(y = 0, color = 'gray', linestyle = '--')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.title("Residuales vs. porcentaje_descuento", fontsize = 14)

# Grafica los residuales vs. 'dias_entrega'
plt.subplot(1, 2, 2)
plt.scatter(df["dias_entrega"], residuales, color = '#008B8B')
plt.xlabel('dias_entrega', fontsize = 12)
plt.ylabel('Residuales', fontsize = 12)
plt.axhline(y = 0, color = 'gray', linestyle = '--')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.title("Residuales vs. dias_entrega", fontsize = 14)
plt.tight_layout()
plt.show()


#9. Realiza la prueba de Shapiro para los residuales
_, valor_p_sh = shapiro(residuales)
print(f'valor-p de Shapiro: {valor_p_sh: 0.4f}\n')

# 10. Realiza la prueba de Breusch-Pagan para los residuales
# Para het_breuschpagan, usa la matriz de diseño del modelo ajustado
exog_for_bp = modelo.model.exog
_, valor_p_bp, _, _ = het_breuschpagan(residuales, exog_for_bp)
print(f'valor_p de Breusch-Pagan: {valor_p_bp: 0.4f}\n')

# 12. Realiza una tabla ANOVA
tabla_anova = sm.stats.anova_lm(modelo)
print("\nTabla ANOVA:\n")
print(tabla_anova)
```

f) Modelo matematico

Usamos una regresión múltiple para predecir el precio en función de todas las variables numéricas y codificadas.
El modelo nos da una ecuación del tipo:

Precio = β₀ + β₁·Costo + β₂·Margen + β₃·Ventas + …

Cada coeficiente (β) nos indica cuánto cambia el precio al variar esa variable, manteniendo las demás constantes.

Por ejemplo:
	•	Si el coeficiente de costo es 1.39, significa que por cada unidad adicional de costo, el precio aumenta en promedio 1.39 unidades.
	•	Si el coeficiente de descuento es negativo, implica que al aumentar el descuento, el precio de venta esperado disminuye.

El R² ajustado fue alto (~0.97), lo que significa que el modelo explica el 97% de la variabilidad del precio.
Eso indica un excelente ajuste, aunque debemos cuidar el sobreajuste.

g) Evaluación del modelo

Comprobando si el modelo cumple los supuestos estadísticos


Hicimos:

	•	Gráficos de residuos vs valores ajustados: la nube de puntos sin patrón indica linealidad.
	•	Prueba de Shapiro-Wilk: dio p < 0.05, lo que sugiere que los residuos no son perfectamente normales, aunque en grandes muestras esto no es grave.
	•	Prueba Breusch-Pagan: no se detectó heterocedasticidad (p > 0.05).
	•	Distancias de Cook: la mayoría de los puntos no son influyentes.
	•	Validación cruzada (CV): mostró un RMSE bajo, confirmando buena precisión y generalización.

Conclusion:
El modelo tiene buen poder explicativo y cumple razonablemente los supuestos.
Solo habría que vigilar la normalidad de los errores y posible sobreajuste por la alta correlación entre variables.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt

# 1. Construye una fórmula OLS que incluya todas las variables categóricas y numéricas
formula_new = 'precio ~ costo + ventas + margen_ganancia + inventario + porcentaje_descuento + dias_entrega + C(categoria) + C(material) + C(color) + C(ubicacion) + C(temporada) + C(marca)'

# 2. Ajusta el modelo OLS
modelo_new = ols(formula_new, data=df).fit()

# 3.Calcula los residuales del nuevo modelo ajustado
residuales_new = modelo_new.resid

# Lista de variables independientes numéricas para graficar
numerical_predictors = ['costo', 'ventas', 'margen_ganancia', 'inventario', 'porcentaje_descuento', 'dias_entrega']

# 4. Crea una figura con una cuadrícula de 2x3 para las gráficas
fig, axes = plt.subplots(2, 3, figsize=(20, 12))
axes = axes.flatten() # Flatten the 2x3 array of axes for easier iteration

#  5. y 6. Genera gráficas de residuales para cada variable independiente numérica
for i, col in enumerate(numerical_predictors):
    ax = axes[i]
    ax.scatter(df[col], residuales_new, color='#008B8B', alpha=0.6)
    ax.set_xlabel(col, fontsize=12)
    ax.set_ylabel('Residuales', fontsize=12)
    ax.axhline(y=0, color='gray', linestyle='--')
    ax.set_title(f'Residuales vs. {col}', fontsize=14)
    ax.spines[['top', 'right']].set_visible(False)

# 7. Ensure tight layout and display the plot
plt.tight_layout()
plt.show()
```
